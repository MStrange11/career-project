```python
import os
```

```python
from keras.preprocessing import image
```

```python
import cv2
```

```python
# https://www.kaggle.com/datasets/belsonraja/face-mask-dataset-with-and-without-mask
```

```python
categories = ['with_mask', 'without_mask']
```

```python
data = []
```

```python
for cate in categories:
    path = os.path.join('train', cate)

    label = categories.index(cate)
    
    for file in os.listdir(path):
        img_path = os.path.join(path, file)
        img = cv2.imread(img_path)
        img = cv2.resize(img,(224,244))
        
        data.append([img, label])
```

```python
data
```

```python
import random
```

```python
import numpy as np
import pandas as pd
```

```python
random.shuffle(data)
```

```python
X= []
Y= []

for features, label in data:
    X.append(features)
    Y.append(label)
```

```python
len(X) == len(Y)
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
print(X.shape)
print(Y.shape)
```

```python
X = X/255
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=.2)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from keras.applications.vgg16 import VGG16
```

```python
vgg = VGG16()
# vgg is a functional model

#  we want sequential
```

```python
vgg.summary()
```

```python
from keras import Sequential
```

```python
model = Sequential()
```

```python
for layer in vgg.layers[:-1]:
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
model.summary()
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics =  ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs= 5, validation_data=(x_test, y_test))
```

```python
import pickle
```

```python
pickle.dump(model, open('face mask.pkl', 'wb'))
```
